```python
# Lets learn setting custom indexes and benefits of doing that
```

```python
people = {
    "first":['vikas','rakesh','mahesh'],
    "last":['sharma','puri','jain'],
    "email":["vikas_sharma@gmail.com","rakesh_puri@gmail.com","mahesh_jain@gmail.com"]
}
```

```python
import pandas as pd
```

```python
df_dict = pd.DataFrame(people)
```

```python
df_dict
```

```python
# Index:

# Indexs are range of values that represents the rows

# Though pandas doesnt restrict indexes to be unique value most of the times they are maintained unqiue
```

```python
# what would be the right index for df_dict data frame????

# Since each person would have a unqiue email id, it would be apt to have email as the index

df_dict.set_index('email')
```

```python
# Note:  The orginal data frame wont change with indexing -- since pandas dont do "in-place" change

df_dict   # still find the default indexing
```

```python
# How to let pandas make the in-place changes i.e. cascade the changes to the original data frame ???

df_dict.set_index('email', inplace=True)
```

```python
df_dict   # now the orginal data frames default indexing also set to 'email'
```

```python
df_dict.index
```

```python
# why indexes are useful ????

# since index acts a unique label to the rows , it helps fetch information in a meaningful way and improves the performance

# instead of saying fetch the information of '0' index, now we may say fetch the information of person with email --- vikas_sharma@gmail.com

df_dict.loc['vikas_sharma@gmail.com']
```

```python
# we can also specify only the speific column of interest  --- thourgh 2nd parameter

df_dict.loc['vikas_sharma@gmail.com', 'last']
```

```python
# post change of indexing to email ------

# loc doesnt support default indexing values ?????  however, iloc still works with default index values ---> df_dict.iloc[0]

df_dict.loc[0]
```

```python
df_dict.iloc[0]
```

```python
#  Can we reset index to default values ???  if got changed accidentally ??

df_dict.reset_index(inplace=True)
df_dict
```
