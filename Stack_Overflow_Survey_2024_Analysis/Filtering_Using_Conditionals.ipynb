```python
people = {
    "first":['vikas','rakesh','mahesh', 'saurabh'],
    "last":['sharma','puri','jain', 'sharma'],
    "email":["vikas_sharma@gmail.com","rakesh_puri@gmail.com","mahesh_jain@gmail.com", "saurabh_sharma@gmail.com"]
}
```

```python
import pandas as pd
```

```python
df_dict = pd.DataFrame(people)
df_dict
```

```python
df_dict['last'] == 'sharma'   # Filter mask --- gives rows meeting filter criterion
```

```python
# Approch 1

filt = ( df_dict['last'] == 'sharma' )   # filter condition

df_dict[filt]                          # return rows from data frame matching the filter condition
```

```python
# Approch 2

df_dict[df_dict['last'] == 'sharma']
```

```python
# Approch 3  ---- using .loc()

df_dict.loc[filt]
```

```python
# select specific columns --- say fetch email id for matching rows

df_dict.loc[filt, 'email']
```

```python
# Note

# In pandas cant use the AND & OR in filter conditions, rather use sybbols --->  '&'  and '|'
```

```python
# filter to return email id,  where last name is 'sharma' & first name is 'vikas'  ????

filt = (df_dict['last'] == 'sharma') & (df_dict['first'] == 'vikas')
df_dict.loc[filt, 'email']
```

```python
# filter to return email id,  where last name is 'sharma' or first name is 'vikas'  ????

filt = (df_dict['last'] == 'sharma') | (df_dict['first'] == 'rakesh')
df_dict.loc[filt, 'email']
```

```python
# Negation ---   '~'

# filter to return email id , where last name is not 'sharma'  or  first name is not 'vikas'

df_dict.loc[~filt,'email']
```

```python
# return rows where salary is > 70000

high_salary_filt = (df['ConvertedComp'] > 70000)
df[high_salary_filt]
```

```python
# For high salaries paid --- fetch their country, computer languages they worked and salary

df.loc[ high_salary_filt, ['Country', 'LanguageWorkedWith', 'ConvertedComp'] ]
```

```python
# Filter the survey results from only 5 countries --- 'United States', India', 'United Kingdom', 'Germany', 'Canada'

countries = ['United States', 'India', 'United Kingdom', 'Germany', 'Canada']    # USE  A  LIST
filt = df['Country'].isin(countries)                                             # mark True or False based on country
df.loc[filt, 'Country']                                                          # return country for rows with True in filter
```

```python
# want to know the people who said they worked with python programming language

# use string method

filt = df['LanguageWorkedWith'].str.contains('Python', na=False)
df.loc[filt, 'LanguageWorkedWith']
```
