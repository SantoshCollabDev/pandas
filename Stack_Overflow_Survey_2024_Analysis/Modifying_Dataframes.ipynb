## 4 popular methods to update multiple rows at once

1. apply
2. map
3. applymap
4. replace

### apply can work on both dataframe or a series but behaviour could be deifferent based on the object type

```python
# apply on series

df_dict['email'].apply(len)
```

```python
def update_email(email):
    return email.upper()

df_dict['email'].apply(update_email)
```

```python
df_dict['email'] = df_dict['email'].apply(update_email)
df_dict
```

```python
# INSTEAD of a seperate/regular function we could have used a lamba  function i.e. a no name function as well

df_dict['email'] = df_dict['email'].apply(lambda x: x.lower())   # can use lambda functions for simple operations or non-complex functions
df_dict
```

```python
# can we apply len function to data frame instead of series ????

df_dict.apply(len)    # instead of len of data frame, It applied the len on each of the series (column) in the data frame
```

```python
len(df_dict['email'])
```

## apply should be used on series objects to apply functions

```python
# if we apply min function on string column it returns first letter in the alphabetical order

df_dict.apply(pd.Series.min)   # min is a pd.Series function
```

```python
df_dict.apply(lambda x: x.min())  # here x is a series i.e.column and this function returns min value from each series
```

## running apply on a series applies the function to every value in the series  &
