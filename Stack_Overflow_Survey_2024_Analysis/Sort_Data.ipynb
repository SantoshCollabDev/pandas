# How to sort data using Pandas

```python
people = {
    "first":['vikas','rakesh','mahesh', 'saurabh'],
    "last":['sharma','puri','jain', 'sharma'],
    "email":["vikas_sharma@gmail.com","rakesh_puri@gmail.com","mahesh_jain@gmail.com", "saurabh_sharma@gmail.com"]
}
```

```python
import pandas as pd
```

```python
df_dict = pd.DataFrame(people)
df_dict
```

```python
# df = pd.read_csv('D:/GIT_Repositories/pandas/survey_results_public.csv', index_col = 'Respondent')
# schema_df = pd.read_csv('D:/GIT_Repositories/pandas/survey_results_schema.csv', index_col = 'Column')

# pd.set_option('display.max_columns', 85)
# pd.set_option('display.max_rows', 85)
```

### sort based on specific column values       ---->  .sort_values( by = 'column name' )

```python
# Sort based on the 'last' name

df_dict.sort_values(by='last')
```

## sort in descending order ---> ascending=False

### .sort_values( by = 'column name' , ascending=False)

```python
df_dict.sort_values( by='last', ascending = False )
```

### sort on multiple columns

#### when 1st column has identical values for rows then sort based on 2nd columns

```python
df_dict
```

####  sort on last name and if duplicates then further sort on first naem


```python

df_dict.sort_values( by = ['last', 'first'] )
```

####  sort on last name in 'DESCENDING' order and first name in 'ASCENDING' order


```python
df_dict.sort_values( by = ['last', 'first'], ascending = [False, True], inplace = True)
```

```python
df_dict
```

### now to restore the original order of indexes -----> .sort_index()

```python
df_dict.sort_index(inplace = True)
```

```python
df_dict
```

### To view only sorted values of a particular column INSTEAD of seeing entire data frame

```python
# Want to see ONLY 'last name' values in sorted order

df_dict['last'].sort_values()
```

## to sort the rows based on country 

```python
df.sort_values(by='Country', inplace=True)
```

```python
df['Country'].head(50)   # check 1st 50 values to see if the rows are re-arranged based on 'country' values
```

## Sort based on country and check the highest salaries in each country

```python
df.sort_values(by=['Country', 'ConvertedComp'], ascending=[True, False], inplace=True)
```

```python
df[['Country', 'ConvertedComp']].head(50)
```

## To grab 10 highest salaries from the surevey results  -----   .nlargest(no_of_values)

```python
df['ConvertedComp'].nlargest(10)
```

# if we want to see entire rows for the 10 highest salaries instead of just salaries column

apply .nlargest on entire dataframe

```python
df.nlargest( 10, 'ConvertedComp' )
```

### to get 10 smallest salaries ---- .nsmallest(10)
