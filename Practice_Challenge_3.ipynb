**MINI CHALLENGE #1:**
- **Define a function that doubles employee salaries and adds $100**
- **Apply the function to the DataFrame**
- **Calculate the updated total Salaries of all employees combined**

```python
employee_df = pd.read_csv('employee_information.csv')
employee_df
```

```python
def double_salary(salary):
    return (2 * (salary)) + 100
```

```python
employee_df['Salary'] = employee_df['Salary'].apply(double_salary)
```

```python
employee_df
```

```python
total_employee_salaries = employee_df['Salary'].sum()
```

```python
print(total_employee_salaries)
```

**MINI CHALLENGE #4:**
- **Assume that you and your college friend become a new employee at the same company and would like to add your first names, last names and unique employee IDs. Define a new DataFrame and add it to the master list "employee_all_df"** 

```python
# Creating a dataframe from a dictionary
# Let's define a dataframe with a list of employees with IDs = 1, 2, 3, 4, 5 
# Check this out: https://pandas.pydata.org/pandas-docs/stable/user_guide/merging.html

employee1_data = {
    'Employee ID': [1, 2, 3, 4, 5],
    'First Name' : ['Nancy', 'Alex', 'Shep', 'Max', 'Allen'],
    'Last Name'  : ['Rob', 'Ali', 'George', 'Mitch', 'Steve']
}

employee1_df = pd.DataFrame( employee1_data, columns = ['Employee ID', 'First Name', 'Last Name'] )

employee1_df
```

```python
# Let's define another dataframe for a separate list of clients (IDs = 6, 7, 8, 9, 10)


employee2_data = {
    'Employee ID': [6, 7, 8, 9, 10],
    'First Name' : ['Bill', 'Dina', 'Sarah', 'Heather', 'Holly'],
    'Last Name'  : ['Christian', 'Mo', 'Steve', 'Bob', 'Michelle']
}

employee2_df = pd.DataFrame( employee2_data, columns = ['Employee ID', 'First Name', 'Last Name'] )

employee2_df
```

```python
new_data = {
    'Employee ID': [11, 12],
    'First Name' : ['Justin', 'Sophie'],
    'Last Name'  : ['Trudeau', 'Trudeau']
}

employee3_df = pd.DataFrame(new_data, columns = ['Employee ID', 'First Name', 'Last Name'] )
employee3_df
```

```python
employee_all_df = pd.concat([employee1_df,employee2_df], ignore_index = True)
employee_all_df
```

```python
employee_all_df = pd.concat( [employee_all_df, employee3_df], ignore_index = True)
employee_all_df
```
